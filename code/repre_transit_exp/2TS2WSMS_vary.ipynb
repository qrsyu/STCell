This script is to plot the sample sensory input and output of 2TS2WSMS experiment which varies the ratio of temporal and spatial channels. 

```python
import numpy as np
from matplotlib import pyplot as plt

load_data_type = f'2TS2WSMS_vary50'
data = np.load(f'../../data/{load_data_type}.npy', allow_pickle=True).item()
test_inputs = data['test_inputs']
test_labels = data['test_labels']
print(test_inputs.shape, test_labels.shape)
temp_reso = 100
```

```python
fig, ax = plt.subplots(4, 2, figsize=(5, 4), sharex=True, sharey=True)
test_inputs[test_inputs == 0] = np.nan
for i, channel in enumerate(range(35, 37)):
    ax[i][0].plot(np.arange(test_inputs.shape[1]) * temp_reso / 1000, test_inputs[0, :, channel], 
               label=f'{i+1}', color='#297270')
    ax[i][0].legend(loc='best')
    ax[i][0].set_xlim(0, test_inputs.shape[1]*temp_reso/1000)
    ax[i][1].plot(np.arange(test_labels.shape[1]) * temp_reso / 1000, test_labels[0, :, channel], 
                label=f'{i+1}', color='#A3432B')
    ax[i][1].legend(loc='best')
    ax[i][1].set_xlim(0, test_labels.shape[1]*temp_reso/1000)
for i, channel in enumerate(range(50, 52)):
    ax[i+2][0].plot(np.arange(test_inputs.shape[1]) * temp_reso / 1000, test_inputs[0, :, channel], 
               label=f'{i+3}', color='#297270')
    ax[i+2][0].legend(loc='best')
    ax[i+2][0].set_xlim(0, test_inputs.shape[1]*temp_reso/1000)
    ax[i+2][1].plot(np.arange(test_labels.shape[1]) * temp_reso / 1000, test_labels[0, :, channel], 
                label=f'{i+3}', color='#A3432B')
    ax[i+2][1].legend(loc='best')
    ax[i+2][1].set_xlim(0, test_labels.shape[1]*temp_reso/1000)
ax[0][0].set_title('Masked Inputs')
ax[0][1].set_title('Outputs')
ax[-1][0].set_xlabel('Time (s)')
ax[-1][1].set_xlabel('Time (s)')
plt.tight_layout()
plt.savefig('2TS2WSMS_exp.png', dpi=300, transparent=True)
```
